```python
Task 4
Create a function that return employees with it's company country.

Expected return:

[{
    "full_name": "John Doe",
    "company": "Company 1",
    "country": "United States"
}, {
    "full_name": "Tom Smith",
    "company": "Company 2",
    "country": "United Kingdom"
}, {
    "full_name": "Andrew Sebastian",
    "company": "Company 3",
    "country": "United Kingdom"
},  {
    "full_name": "Ying Han Tan",
    "company": "Company 1",
    "country": "United States"
},  {
    "full_name": "Kenneth Ng",
    "company": "Company 3",
    "country": "United States"
},  {
    "full_name": "Rubby Lee",
    "company": "Company 2",
    "country": "United Kingdom"
},   {
    "full_name": "Robert White",
    "company": "Company 1",
    "country": "United States"
},
    
]
```

```python
def get_employees_with_company_country(employee_list, company_list):
    result = []
    
    # Create a mapping of company names to their countries
    company_country_map = {company['name']: company['country'] for company in company_list}
    
    # Build the result list
    for employee in employee_list:
        full_name = employee['full_name']
        company_name = employee['company']
        
        # Get the corresponding country from the mapping
        country = company_country_map.get(company_name, "Unknown")
        
        result.append({
            "full_name": full_name,
            "company": company_name,
            "country": country
        })
    
    return result

# Call the function and print the result
employees_with_country = get_employees_with_company_country(employee_detail_list, company_detail_list)
print(employees_with_country)
```

```python
Task 5
Create a function that return companies with a number of employees nationality that are working in the company.

Expected return:

[{
    "company": "Company 1",
    "employee_nationality": {
        "Australia": 2,
        "Hong Kong": 1
    }
}, {
    "company": "Company 2",
    "employee_nationality": {
        "Hong Kong": 1,
        "United States": 1
    }
}, {
    "company": "Company 3",
    "employee_nationality": {
        "United States": 2
    }
}]
```

```python
def get_companies_with_employee_nationality(employee_list, company_list):
    result = []
    
    # Create a mapping of company names to their employee nationalities
    company_nationality_map = {}
    
    # Populate the mapping
    for employee in employee_list:
        company_name = employee['company']
        nationality = employee['nationality']
        
        if company_name not in company_nationality_map:
            company_nationality_map[company_name] = {}
        
        if nationality not in company_nationality_map[company_name]:
            company_nationality_map[company_name][nationality] = 0
        
        company_nationality_map[company_name][nationality] += 1

    # Build the result list
    for company in company_list:
        company_name = company['name']
        employee_nationality = company_nationality_map.get(company_name, {})
        
        result.append({
            "company": company_name,
            "employee_nationality": employee_nationality
        })
    
    return result

# Call the function and print the result
companies_with_nationality = get_companies_with_employee_nationality(employee_detail_list, company_detail_list)
print(companies_with_nationality)
```
